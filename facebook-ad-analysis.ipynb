# **Computing Number of impressions in Age range 30-34**

```python
ageRestrictedDataset =dataset[dataset.age == "30-34"] 
totalImpressions = ageRestrictedDataset.impressions.sum()
print(f"Total Impressions in age between 30-34 are : {totalImpressions}.")
```

**Total Impressions in age between 30-34 are : 35678593.0**

### **converting reporting_start from string to Date and time.** 

```python
from dateutil import parser
for x in range(0,len(dataset)):
    dataset.reporting_start[x] = parser.parse(dataset.reporting_start[x])
dataset
```

### **Computing Number of clicks from date range 19/08/2017 - 22/08/2017**

```python
from dateutil import parser
startDate = parser.parse('19/08/2017')
endDate = parser.parse('22/08/2017')
dateRestrictedDataset = dataset[dataset.reporting_start<= endDate]
len(dateRestrictedDataset)
# dateRestrictedDataset
dateRestrictedDataset = dateRestrictedDataset[dateRestrictedDataset.reporting_start>=startDate]
print(f"total clicks in date range 19/08/2017 - 22/08/2017 are : {dateRestrictedDataset.clicks.sum()}")
```

**total clicks in date range 19/08/2017 - 22/08/2017 are : 3763**
